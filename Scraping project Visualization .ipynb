```python
approvalrate = pd.read_csv("~/Desktop/ObamaTrump/approvalrate.csv")
meanincome = pd.read_csv("~/Desktop/ObamaTrump/meanincome.csv")
povertyrate = pd.read_csv("~/Desktop/ObamaTrump/povertyrate.csv")
policemurderrate = pd.read_csv("~/Desktop/ObamaTrump/policemurderrate.csv")
```

```python
import math
meanincome.meanincome = meanincome.meanincome
for i in range(len(meanincome.meanincome)):
    meanincome.meanincome[i] = meanincome.meanincome[i].replace(',','')
```

```python
meanincome.meanincome = meanincome.meanincome.astype(int)
meanincome.meanincome = meanincome.meanincome.apply(lambda x: x/500)
```

```python
approval = {"x": sorted( approvalrate.year.unique()), 
          "y": approvalrate.approvalrate.tolist(), 
          "marker": {"color": "pink", "size": 12}, 
          "mode": "markers", 
          "name": "Approval rate", 
          "type": "scatter"
}

meanin = {"x": sorted(meanincome.year.unique()), 
          "y": meanincome.meanincome.tolist(), 
          "marker": {"color": "blue", "size": 12}, 
          "mode": "markers", 
          "name": "mean income/500", 
          "type": "scatter"
}

data = [approval, meanin]
layout = {"title": "correlation between approval rates of Obama/Trump and mean income of black people 2009-2017", 
          "xaxis": {"title": "years", }, 
          "yaxis": {"title": "apporval rate/income"}}

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filenmae='basic_dot-plot')
```
